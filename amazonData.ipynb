```python
# _INTENTS[]
# _SCENARIOS[]

def process_row(mongolian, english):
    instruction = f"Translate this sentence into Mongolain => {english['sentence']} "
    answer = f"Mongolian translation => {mongolian['sentence']}"
    return {"instruction": instruction, "answer":answer}

def create_new_list(dataset, datasetEnglish):
    # This function iterates over the dataset and applies process_row to each row,
    # collecting the results in a new list.
    # new_list = [process_row(row) for row, english in dataset, datasetEnglish]
    new_list = [process_row(mongolian, english) for mongolian, english in zip(dataset['dev'], datasetEnglish['dev'])]
    new_listDev = [process_row(mongolian, english) for mongolian, english in zip(dataset['devtest'], datasetEnglish['devtest'])]
    return new_list + new_listDev

# Example dataset

# Create the new list
new_translation_list = create_new_list(datasetMongolian, datasetEnglish)

# Output the new list
# print(new_list)
```

```python
new_translation_list
```

```python
import jsonlines
import math

# Determine the index at which to split the data
split_index = math.ceil(len(new_translation_list) * 0.8)

# Split the data into training and validation sets
train_data = new_translation_list[:split_index]
validation_data = new_translation_list[split_index:]

# Save the training data to train.jsonl
with jsonlines.open('train.jsonl', mode='a') as writer:
    writer.write_all(train_data)

# Save the validation data to validation.jsonl
with jsonlines.open('valid.jsonl', mode='a') as writer:
    writer.write_all(validation_data)
```

```python
beleDataset = load_dataset("facebook/belebele", "khk_Cyrl")
```

```python
beleDataset['test'][0]
```

```python
# _INTENTS[]
# _SCENARIOS[]

def process_row(data):
    instruction = f"Дараах өгөгдөл дээр суурилан асуултанд зөв хариултаар хариулна уу? Өгөгдөл => {data['flores_passage']} \n Асуулт => {data['question']} \n Боломжит хариултууд 1){data['mc_answer1']} 2){data['mc_answer2']} 3){data['mc_answer3']} 4){data['mc_answer4']}"
    correct_answer = 'mc_answer' + data['correct_answer_num']
    answer = f"Зөв хариулт бол {data['correct_answer_num']}){data[correct_answer]}"
    return {"instruction": instruction, "answer":answer}

def create_new_list(dataset):
    new_list = [process_row(data) for data in dataset['test']]
    return new_list

# Example dataset

# Create the new list
correctAnswer = create_new_list(beleDataset)

# Output the new list
# print(new_list)
```

```python
correctAnswer
```

```python
# Determine the index at which to split the data
split_index = math.ceil(len(correctAnswer) * 0.8)

# Split the data into training and validation sets
train_data = correctAnswer[:split_index]
validation_data = correctAnswer[split_index:]

# Save the training data to train.jsonl
with jsonlines.open('train.jsonl', mode='a') as writer:
    writer.write_all(train_data)

# Save the validation data to validation.jsonl
with jsonlines.open('valid.jsonl', mode='a') as writer:
    writer.write_all(validation_data)
```
