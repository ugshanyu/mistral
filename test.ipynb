```python
from transformers import pipeline

# Load the model and tokenizer from Hugging Face
generator = pipeline('text-generation', model='ugshanyu/llama-2-7b-miniguanaco')

# Generate text
text = generator('Translate this Mongolian sentence to English (Амьдрал сайхан)', max_length=50, do_sample=True, temperature=0.7)

# Output the generated text
print(text[0]['generated_text'])
```
